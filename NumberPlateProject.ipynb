```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('D:ciaz.jpg')
```

```python
def display(img):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

```python
display(img)
```

```python
carPlate = cv2.CascadeClassifier('D:/DATA/haarcascades/haarcascade_russian_plate_number.xml')
```

```python
def detect_plate(img):
    car = img.copy()
    plateRect = carPlate.detectMultiScale(car,1.4,3)
    for(x,y,w,h) in plateRect:
        cv2.rectangle(car,(x,y),(x+w,y+h),(0,0,255),5)
        print(x,y,w,h)
    return car
```

```python
result = detect_plate(img)
```

```python
display(result)
```

```python

x,y,w,h = [70,72,118,39]
def detect_and_blur_plate(img):
    img1 = img.copy()
    car = img.copy()
    img1=img[y:y+h,x:x+w]
    blurimg = cv2.medianBlur(img1,7) 
    car[y:y+h,x:x+w] = blurimg
    return car
    
    
```

```python
result = detect_and_blur_plate(img)
```

```python
display(result)
```
